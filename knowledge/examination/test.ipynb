```python
import pandas as pd

df = pd.read_json("./procedural_examination_chapters.json")

MAP = {
    "一": "1", "二": "2", "三": "3", "四": "4", "五": "5",
    "六": "6", "七": "7", "八": "8", "九": "9", "十": "10",
    "十一": "11", "十二": "12", "十三": "13",
    "十四": "14",
    "十五": "15",
    "十六": "16",
    "十七": "17",
    "十八": "18",
    "十九": "19",
    "二十": "20",
    "二十一": "21",
    "二十二": "22",
    "二十三": "23",
}

dct = {}
for i in df["chapter"].unique().tolist():
    j = i.split(" ")[1].replace("第", "").replace("章", "")

    dct[i] = MAP[j]

dct
```

```python
# df["chapter"] = df["chapter"].map(dct)

import os
os.makedirs("./01", exist_ok=True)
for i in dct.keys():
    tmp = df.loc[df["chapter"] == i, :]
    name = dct[i].zfill(2)
    tmp.to_json(f"./01/1-{name}.json", orient="records", force_ascii=False, indent=4)
```
